# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

import os
os.chdir("C:/Users/AFOS-Pavilion-20/Documents/Naive Bays Algorithm")
```

```python
data = pd.read_csv("golfdataset.csv")
data
```

# Encoding data : string to numerical

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Outlook'] = le.fit_transform(data.Outlook.values)
data['Temp'] = le.fit_transform(data.Temp.values)
data['Humidity'] = le.fit_transform(data.Temp.values)
data['Windy'] = le.fit_transform(data.Windy.values)
data['PlayGolf'] = le.fit_transform(data.PlayGolf.values)
```

```python
data
```

```python
# Splitting data into train and test
labels = data['PlayGolf']
features = data.drop('PlayGolf', axis=1)

x_train,  x_test, y_train, y_test = train_test_split(features, labels, test_size=0.4, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Training the model on the training set
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
# Making prediction on the testing set
y_pred = gnb.predict(x_test)
#y_pred = gnb.predict([[2,0,1,0]])

y_pred
```

```python
# Comparing actual response values (y_test) with predicted response (y_pred) 
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```
